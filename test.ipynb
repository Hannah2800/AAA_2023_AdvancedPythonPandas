```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_trip_counts(data):
    # Convert 'Trip_Start_Timestamp' column to datetime
    data['Trip_Start_Timestamp'] = pd.to_datetime(data['Trip_Start_Timestamp'])

    # Group by hour and count the number of trips
    trip_counts = data.groupby(data['Trip_Start_Timestamp'].dt.hour)['Trip_ID'].count()

    # Plot the results
    plt.figure(figsize=(10, 6))
    trip_counts.plot(kind='bar', color='blue')
    plt.xlabel('Hour of the Day')
    plt.ylabel('Number of Trips')
    plt.title('Trip Counts by Hour')
    plt.xticks(rotation=0)
    plt.show()
```

```python
# Assuming your data frame is named 'df'
plot_trip_counts(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_trip_counts_4(data):
    # Convert 'Trip_Start_Timestamp' column to datetime
    data['Trip_Start_Timestamp'] = pd.to_datetime(data['Trip_Start_Timestamp'])

    # Group by hour and minute and count the number of trips
    trip_counts = data.groupby([data['Trip_Start_Timestamp'].dt.hour, data['Trip_Start_Timestamp'].dt.minute])['Trip_ID'].count()

    # Plot the results
    plt.figure(figsize=(18, 6))
    trip_counts.plot(kind='bar', color='blue')
    plt.xlabel('Hour and Minute of the Day')
    plt.ylabel('Number of Trips')
    plt.title('Trip Counts by Hour and Minute')
    plt.xticks(rotation=90)
    plt.show()
```

```python
plot_trip_counts_4(df_chicago)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import timedelta

def round_to_nearest_hour(timestamp):
    # Round the timestamp to the nearest hour
    rounded_timestamp = timestamp.round('H')
    return rounded_timestamp

def plot_trip_counts_rounded(data):
    # Convert 'Trip_Start_Timestamp' column to datetime
    data['Trip_Start_Timestamp'] = pd.to_datetime(data['Trip_Start_Timestamp'])

    # Round the timestamps to the nearest hour
    data['Rounded_Timestamp'] = data['Trip_Start_Timestamp'].apply(round_to_nearest_hour)

    # Group by rounded hour and count the number of trips
    trip_counts = data.groupby(data['Rounded_Timestamp'])['Trip_ID'].count()

    # Plot the results
    plt.figure(figsize=(10, 6))
    trip_counts.plot(kind='bar', color='blue')
    plt.xlabel('Rounded Hour')
    plt.ylabel('Number of Trips')
    plt.title('Trip Counts (Rounded to Nearest Hour)')
    plt.xticks(rotation=0)
    plt.show()
```

```python
plot_trip_counts_rounded(df_chicago)
```
